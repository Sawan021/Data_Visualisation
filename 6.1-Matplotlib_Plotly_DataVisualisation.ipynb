## MATPLOTLIB: Visualisation with Python

DATA VISUALISATION <br>
It is used to give intermediate level data understanding based on some patterns in data.Data visualization is the representation of data through use of common graphics, such as charts, plots, infographics, and even animations. <br>
These visual displays of information communicate complex data relationships and data-driven insights in a way that is easy to understand. 
(1)D.V is generally done for internal analysis to get some useful insights into the data.
(2) Used for experimental purposes specially in case when ML algorithms are used.

Dashboarding is different as it is generation of final insight of data like a report, which is generally done at the end of a project.
It is created for external analysis or presentations.

Histogram: Represents occurences of a Range (x-axis) with its frequency (y-axis)
Bar Graph: Represents categorical data using rectangular bars where the length of each bar is proportional to the value they represent. 

```python
df1['sepal_length'].plot.hist()
```

```python
df1.plot(kind='hist',stacked=True, bins=100,figsize = (10,10) , title = "histogram of iris")

# bins is width
```

```python
df1.plot(kind = 'hist' , stacked = True , bins = 100 , figsize = (10,10) , title = "histogram of iris" ,orientation = 'horizontal')
```

```python
df1.hist(figsize= (10,10) , color = 'r')
```

```python
df1.hist(figsize= (10,10) , color = '#66ff99')
```

```python
df1.hist(figsize= (10,10) , color = 'r' , alpha  = .3)

#alpha contrast increasing as move towards 1.
```

```python
# understanding relationship between parameters SCATTER PLOT is used

# in case of this plot we need to set axis

df1.plot.scatter(x='sepal_width',y='petal_width')
```

```python
df1.plot.scatter(x='sepal_length',y='petal_width')
```

```python
df1.plot.scatter(x='sepal_width',y='petal_width', c='petal_length')

#Here we are seeing relationship b/w x and y with color variation due to c
```

```python
df1.plot.scatter(x='sepal_length',y='sepal_width',label='skm',s=60,color='g')

#size controls radius of these dots
```

```python
df1.plot.hexbin(x='sepal_length',y='sepal_width',label='skm',gridsize=7,color='g')
```
