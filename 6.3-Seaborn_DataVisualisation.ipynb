## Seaborn: Statistical Data Visualization

Its a library used for plotting graphs in python. We can consider it has an high level API of matplotlib as its built over matplotlib and also integrated with pandas data structure.

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips)
```

```python
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips,hue='smoker')
#hue helps separation based on categorical data
```

```python
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips,hue='day')
```

```python
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips,hue='day',style='smoker')
```

```python
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips,hue='day',style='smoker', size='size')
```

```python
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips ,size = 'tip')
```
